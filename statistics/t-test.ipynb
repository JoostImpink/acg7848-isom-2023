# T-test (or Student t-test)



The t-test is used to do a statistical test (hypothesis testing) for one or two groups of data. (You need regression analysis if there are three groups or more.)

There are three different t-tests:
- One sample t-tests: there is one series (one mean). Example: do students drink more than 10 beers per week?
- Two sample t-tests: there are two series (two means) (independent measures) Example: do men drink more beer than women?
- Paired t-test: the difference between two series (dependent measures) Example: do men drink more beer on Saturday than Sunday?

See also: https://www.jmp.com/en_us/statistics-knowledge-portal/t-test.html

## One sample t-test

The t-value is the difference (mean vs some target value, the target value can be 0) divided by the standard error of the mean.

Standard deviation ($\sigma$): $ \sigma = \sqrt{\frac{1}{N-1} \sum_{i=1}^N (x_i - \overline{x})^2}$
    
Standard error of the mean (SEM) = $ \frac{\sigma}{\sqrt{n}}   $

So, for example, to test if the mean ($\bar{x}$) is positive (>0):
    
$ t = \frac{\bar{x}\sqrt{n}}{\sigma} $

## Two sample t-test

In a two sample t-test the difference in means between the two groups is divided by the standard error of the difference.

If both groups have the same variance:

$ t=\frac{\bar{x}_1-\bar{x}_2}{\displaystyle
\biggl[
\frac{(N_1-1)\hat{\sigma}_1^2 + (N_2-1)\hat{\sigma}_2^2}{N_1+N_2-2}
\biggl(\frac{1}{N_1}+\frac{1}{N_2}\biggr)
\biggr]^{1/2}}$

From: https://tex.stackexchange.com/questions/456051/standard-deviation

If both groups have different variances:

$ t = \frac{\bar{x}_1 - \bar{x}_2 }{\sqrt{s^2_1/n_1+s^2_2/n_2  }} $
where $s= \frac{\sigma}{\sqrt{n}} $

## Paired t-test

The match for the paired t-test is the same as for the one sample t-test:

$t = \frac{Y\sqrt(n)}{\sigma}$ where Y is the difference in scores (for example: the average beers on Sunday minus the average beers on Saturday).

## Example paired t-test 

Let's read the glassdoor review data where employees have left ratings on work-life balance, compensation, etc.

Let's test if compensation has higher ratings than work-life balance.

```python
import pandas as pd
import scipy.stats as stats
```

```python
df = pd.read_csv('../datasets/glassdoor.csv')
df.head()
```

```python
help(stats.ttest_rel)
```

```python
# compare compensation rating with worklife rating
# nan_policy = 'omit' ignores missing values, otherwise output is 'nan' because some values are missing
t_statistic, p_value = stats.ttest_rel(df['compensation'] , df['worklife'], nan_policy = 'omit')
# print output
print(t_statistic , p_value)
# a high value for the statistic (and a low p-value) reject the null hypothesis of no difference
```

### One sample t-test

```python
help(stats.ttest_1samp)
```

```python
# compute the difference between compensation and worklife
diff = df['compensation'] - df['worklife']
# test if it is statistically significantly different from 0
t_statistic, p_value = stats.ttest_1samp(a=diff, popmean=0, nan_policy = 'omit')
# print output
print(t_statistic , p_value)
```
